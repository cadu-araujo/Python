```python
def reverso(video):
    cap = cv2.VideoCapture(video)
    while True:
        rval, frame = cap.read()
        if not rval:
            break
        lista.append(frame)
        lista.reverse()

    for l in lista:
        cv2.imshow('frame', l)

        key = cv2.waitKey(100) & 0xFF
        if key == ord('q'):
            break
        
    cap.release()
    cv2.destroyAllWindows()
```
